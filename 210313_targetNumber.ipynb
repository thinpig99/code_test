이런식으로 깊이를 계속 타고 내려가면서 일정 숫자가 완성되면 개수를 더하고,

스택에서 버린 다음 다음 노드로 가는 것이다.

```python
import numpy as np
```

```python
numbers = [1, 1, 1, 1, 1]
target = 3
```

```python
graph = []
for i in numbers:
    graph.append([i, -i])
```

```python
visited = np.zeros((len(graph), 2), dtype=bool).tolist()
```

```python
graph, visited
```

```python
def dfs(graph, start, visited):
    global target
    global number
    global count

    # 아래 층의 왼쪽 방이 False 면 True로 바꾸고, 내려간다.
    if visited[start + 1][0] == False:
        visited[start + 1][0] = True
        # True로 바꿀 때마다 더해준다.
        number.append(graph[start + 1][0])
        # 층수를 낮춰준다.
        start += 1

        print("왼쪽 방 진입", "-"*70)
        print("방문 이력, visited: ", visited)
        print("현재 층, start: ", start)
        print("현재 number : ", number, np.sum(number))

        # 지하로 내려갈 수 없을 때는 옆방을 체크하러 간다.
        if start == len(graph) - 1:
            print("지하 도착")
            # 지하에 닿았다는 것은 그 경우의 끝까지 간 것을 의미
            # count를 계산해준다.
            if np.sum(number) == target:
                count += 1
                print("지하 도착, count는?: ", count)

            number.pop() # 위층으로 올라가기 전에 수를 빼준다.
            start -= 1
            dfs(graph, start, visited)
        else:
            print("아래층으로")
            dfs(graph, start, visited)

    # 1-2 왼쪽 게 True면 오른쪽을 점검한다.
    # 왼쪽만 True고 오른쪽은 False면
    elif visited[start + 1][1] == False:
        visited[start + 1][1] = True
        # True로 바꿀 때마다 더해준다.
        number.append(graph[start + 1][1])
        # 층수를 낮춰준다.
        start += 1

        print("오른쪽 방 진입", "-"*70)
        print("방문 이력, visited: ", visited)
        print("현재 층, start: ", start)
        print("현재 number : ", number, np.sum(number))

        # 지하로 내려갈 수 없을 때는 옆방을 체크하러 간다.
        if start == len(graph) - 1:
            print("지하 도착")
            # 지하에 닿았다는 것은 그 경우의 끝까지 간 것을 의미
            # count를 계산해준다.
            if np.sum(number) == target:
                count += 1
                print("지하 도착, count는?: ", count)

            number.pop() # 위층으로 올라가기 전에 수를 빼준다.
            start -= 1
            dfs(graph, start, visited)
        else:
            print("아래층으로")
            dfs(graph, start, visited)

    # 둘 다 True면 이미 밑을 다 썼다는 것으로 위로 올라간다.
    # 올라갈 때는 자기 자신을 초기화하고 올라간다.
    else:
        print("양쪽이 모두 True 입니다!")
        print("현재 층: ", start)
        print("현재 이력: ", visited)
        print("현재 number : ", number, np.sum(number))

        # 옥상으로 올라갈 수 없을 때는 건물 밖을 나간다.
        if start == -1:
            pass
        else:
            # 위층으로 올라가기 전에 수를 빼준다.
            number.pop()

            # 올라가기 전 해당 층을 초기화 해준다.
            visited[start + 1][0] = False
            visited[start + 1][1] = False

            # 한 층 올라간다.
            start -= 1
            print("위로 진출", "**"*70)
            print("현재 이력, visited: ", visited)
            print("현재 층, start: ", start)
            print("현재 number : ", number)
            dfs(graph, start, visited)

    print("모든 경우를 수색했음.")
    print(count)
```

```python
def solution(numbers, target):
    graph = []
    for i in numbers:
        graph.append([i, -i])
    visited = np.zeros((len(graph), 2), dtype=bool).tolist()

    start = -1 # 시작 노드의 층
    number = [] # target number
    count = 0 # target number count
    
    dfs(graph, start, visited)

    print(count)
    return count
```
